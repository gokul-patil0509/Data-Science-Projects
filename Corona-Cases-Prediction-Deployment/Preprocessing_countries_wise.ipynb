```python
import pandas as pd
import numpy as np
import pickle


### Loading the dataset
df = pd.read_csv('F:\gokul\covid19_contries_wise.csv')
```

```python
df
```

```python
date=df['Date'].str.split('-',expand=True)
date.columns=['Day', 'Month', 'Year']
date.head()
```

```python
df = pd.concat([df,date],axis=1)
df
```

```python
Final_Dataset = df.groupby(['Day', 'Month', 'Year', 'Country', 'Current_Status'])[['Total_Cases']].sum()
Final_Dataset
```

```python
Final_Dataset.to_csv('Countries_wise.csv')
```
